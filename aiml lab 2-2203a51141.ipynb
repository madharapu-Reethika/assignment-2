```python
import pandas as pd
import matplotlib.pyplot as plt
people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]
health = pd.DataFrame({'People': people, 'Age': age, 'Height': height, 'Weight': weight})
#scatter
plt.figure(figsize=(8, 6))
plt.scatter(health['Age'], health['Weight'])
plt.title('Scatter Plot of Age vs Weight')
plt.xlabel('Age')
plt.ylabel('Weight')
plt.grid()
plt.show()

# Bar Chart
plt.figure(figsize=(8, 6))
plt.bar(health['People'], health['Height'], color='green')
plt.title('Bar Chart of People vs Height')
plt.xlabel('People')
plt.ylabel('Height')
plt.grid(axis='y')
plt.show()

#  Histogram
plt.figure(figsize=(8, 6))
plt.hist(health['Age'],color='pink')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```
